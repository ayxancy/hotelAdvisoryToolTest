Stop Words


```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
```


Lemmatization

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
input_str="books wants card women"
input_str=nltk.word_tokenize(input_str)
for word in input_str:
  print(lemmatizer.lemmatize(word))
```

Applying


```python
#Positive review preprocessing
for index,row in df.iterrows():
  filter_sentence = []
  sentence=df['Positive_Review'][index]
  sentence=re.sub(r'[^\w\s]','', sentence)
  words = nltk.word_tokenize(sentence)
  words= [w for w in words if not w in stop_words]
  for word in words:
    filter_sentence.append(lemmatizer.lemmatize(word))
  print(filter_sentence)
  df.loc[index,'Positive_Review'] = ' '.join(filter_sentence)
  if(index==10000):
    break
```

```python
#Negative review preprocessing
for index,row in df.iterrows():
  filter_sentence = []
  sentence=df['Negative_Review'][index]
  sentence=re.sub(r'[^\w\s]','', sentence)
  words = nltk.word_tokenize(sentence)
  words= [w for w in words if not w in stop_words]
  for word in words:
    filter_sentence.append(lemmatizer.lemmatize(word))
  print(filter_sentence)
  df.loc[index,'Negative_Review'] = ' '.join(filter_sentence)
  if(index==10000):
    break
```

```python
# importing the pipeline module
from transformers import pipeline

# Downloading the sentiment analysis model
SentimentClassifier = pipeline("sentiment-analysis")
```

Sentiment analysis model finds out if your review is negative or positive sentiment. Lets test it

```python
#sentiment analysis function testing
SentimentClassifier(["I love cats so much!",
                     "We had a horrible experience in this trip",
                     "Houston we don't have a problem"
                      ])
```

```python
def FunctionBERTSentiment(inpText):
  return(SentimentClassifier(inpText)[0]['label'])
```

```python
df['Combined_Review'] = df['Positive_Review'] + ' ' + df['Negative_Review']
```

```python
df['Sentiment_Review'] = df['Combined_Review'].apply(FunctionBERTSentiment)
```

```python
df.head(15)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming you have a DataFrame df with 'Positive_Review' and 'Negative_Review' columns
df['sentiment'] = df.apply(lambda row: 1 if row['Sentiment_Review'] == 'POSITIVE' else 0 , axis=1)

# Combine text from positive and negative columns into a single column
df['combined_text'] = df['Positive_Review'] + ' ' + df['Negative_Review']

# Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X = tfidf_vectorizer.fit_transform(df['combined_text'])

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, df['sentiment'], test_size=0.2, random_state=42)
```

```python
df.head(10)
```

```python
# Model selection and training
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, y_train)

# Evaluation
y_pred = logistic_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Deployment: Deploy the model for making predictions on new data
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression


model = logistic_model

def recommend_hotel(comment):
    # Vectorize the input comment
    comment_vectorized = tfidf_vectorizer.transform([comment])

    # Predict sentiment
    sentiment = model.predict(comment_vectorized)

    # Based on sentiment, recommend a hotel
    if sentiment[0] == 1:
        recommended_hotel = df.loc[df['sentiment'] == 1].sample(1)
    else:
        recommended_hotel = df.loc[df['sentiment'] == 0].sample(1)

    return recommended_hotel['Hotel_Name'].values[0]

# Get input comment from the user
user_comment = input("Enter your comment: ")

# Get a hotel recommendation based on sentiment
recommended_hotel = recommend_hotel(user_comment)

# Display the recommendation to the user
print(f"Recommended Hotel: {recommended_hotel}")
```
